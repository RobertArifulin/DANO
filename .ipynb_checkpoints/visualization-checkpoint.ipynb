# Полезные функции

```python
def remove_fliers(df, colums):
    df_clean = df
    for c in colums:
        Q1 = df_clean[c].quantile(q=.25) # квартели
        Q3 = df_clean[c].quantile(q=.75)
        IQR = Q3 - Q1  # Межквартильный размах  
        df_clean = df_clean[(df_clean[c] > (Q1-1.5*IQR)) & ((df_clean[c] < (Q3+1.5*IQR)))] 
        # Фильтрация данных
    return df_clean
```

```python
#функции для аппроксимации 
def pol3(x, a, b, c, d):
    return x**3 * a + x**2 * b  + x * c  + d

def pol1(x, a, b):
    return x * a  + b

def hyp(x, a, b):
    return 1 / (x * a  + b)

def ahyp(x, a, b):
    return 1 / (-x * a) + b

def corr(s):
    return f"Корреляция между велечинами = {round(s.corr().iloc[1, 0], 3)}"
```

```python
def add_rating(df):
    a = df
    a["ratings_cnt_ya"].fillna(0, inplace=True)
    a["reviews_cnt_yell"].fillna(0, inplace=True)
    a["reviews_cnt_2gis"].fillna(0, inplace=True)
    a["reviews_cnt_avito"].fillna(0, inplace=True)

    a["rating_ya"].fillna(0, inplace=True)
    a["rating_yell"].fillna(0, inplace=True)
    a["rating_2gis"].fillna(0, inplace=True)
    a["rating_avito"].fillna(0, inplace=True)
    a["reviews_cnt"] = a["ratings_cnt_ya"] + a["reviews_cnt_2gis"] + a["reviews_cnt_yell"] +a["reviews_cnt_avito"]
    a["reviews_cnt"].replace(0, 1, inplace=True)
    a = a.dropna(subset=['ratings_cnt_ya', 'reviews_cnt_yell', "reviews_cnt_2gis", "reviews_cnt_avito"], how='all')
    a["rating"] = (df["rating_ya"] * df["ratings_cnt_ya"] + df["rating_2gis"] * df["reviews_cnt_2gis"] + df["rating_yell"] * df["reviews_cnt_yell"] + df["rating_avito"] * df["reviews_cnt_avito"])/df["reviews_cnt"]
    a[a["reviews_cnt"] == 0]["rating"] = a[a["reviews_cnt"] == 0][["rating_ya", "rating_yell", "rating_2gis", "rating_avito"]].max(axis=1)
    a = a[(a["rating"] > 0) & (a["rating"] < 5)]
    return a
```

```python
def add_trust_1(df):
    """
    "trust" = (1 / "mvd_request_num" + 1) + (100 - "phb_bad_perc") / 100 + 1 / ("mvd_request_num" + 1) * (100 - "phb_bad_perc") / 100) / 3
    """
    a = df
    a["mvd_request_num"].fillna(0, inplace=True)
    a["trust"] = (1 / (a["mvd_request_num"] + 1) + (100 - a["phb_bad_perc"]) / 100 + 1 / (a["mvd_request_num"] + 1) * (100 - a["phb_bad_perc"]) / 100) / 3
    return a
```

```python
def add_trust_2(df):
    """
    "mvd_request_num" - normalization 
    "trust" = (1 / "mvd_request_num" + 1) + (100 - "phb_bad_perc") / 100 + 1 / ("mvd_request_num" + 1) * (100 - "phb_bad_perc") / 100) / 3
    """
    a = df
    a["mvd_request_num"] = a["mvd_request_num"].fillna(0)
    a = a[(a["mvd_request_num"] > 0) & (a["mvd_request_num"] < 20)]

    a = a[(a["phb_bad_perc"] < 0.1) & (a["phb_bad_perc"] > 0)]

    a["phb_bad_perc"] = (a["phb_bad_perc"] - a["phb_bad_perc"].min())/(a["phb_bad_perc"].max() - a["phb_bad_perc"].min())
    # a["mvd_request_num"] = (a["mvd_request_num"] - a["mvd_request_num"].min())/(a["mvd_request_num"].max() - a["mvd_request_num"].min())
    a["trust"] = (1 / (a["mvd_request_num"] + 1) + (1 - a["phb_bad_perc"]) + (1 / (a["mvd_request_num"] + 1)) * (1 - a["phb_bad_perc"])) / 3
    return a
```

## appl_num phb_all

```python
df = pd.pandas.read_excel(r'data/tinkoff_legal_entities_new.xlsx', index_col="id")

colums = ["appl_num", "phb_all"]
df_clean = remove_fliers(df, colums)
```

```python
fig3, ax4 = plt.subplots(1, 1,  figsize=(12, 6))

x = "appl_num"
y = "phb_all"

a = pd.DataFrame()
a[y] = df_clean[y]
a[x] = df_clean[x]
a = a.dropna()
ax4.scatter(a[x], a[y], s=30, alpha=0.5, zorder=2)

ax4.set_xlabel("Количество заявок на кредит", size=16)
ax4.set_ylabel("Количество телефонов", size=16)

popt, pcov = curve_fit(pol1, a[x], a[y])
plt.plot(a[x].sort_values(), pol1(a[x].sort_values(), *popt), color="green", label="Линейная аппроксимация", zorder=1) 
plt.legend(prop={'size': 16}) #, framealpha=1.0, facecolor="white")


fig3, ax4 = plt.subplots(1, 1,  figsize=(12, 6))
a_mean = a.groupby(by=x).mean()
a_median = a.groupby(by=x).median()
plt.bar(a_mean.index, a_mean[y], label="Среднее занчение")
plt.bar(a_median.index, a_median[y], label="Медианное значение")
ax4.set_xlabel("Количество заявок на кредит", size=16)
ax4.set_ylabel("Количество телефонов", size=16)
plt.legend(prop={'size': 16})
print(corr(a))
```
